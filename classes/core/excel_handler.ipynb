## Reordenar hojas

```python
import re
import time
import os
import win32com.client as win32

try:
    # Iniciar Excel
    excel = win32.Dispatch('Excel.Application')
    excel.Visible = False  # Ejecutar Excel en segundo plano

    # Ruta del archivo original

    # Abrir el libro de trabajo original
    original_wb = excel.Workbooks.Open(file_path)

    # Crear un nuevo libro de trabajo donde moveremos las hojas
    new_wb = excel.Workbooks.Add()
    while new_wb.Sheets.Count > 1:
        new_wb.Sheets(1).Delete()  # Eliminar hojas predeterminadas

    # Directorio de checkpoints
    os.makedirs(save_dir, exist_ok=True)  # Crear carpeta de checkpoints si no existe

    # Extraer nombres y ordenarlos
    sheets_info = []
    for sheet in original_wb.Sheets:
        name = sheet.Name.lower()
        match = re.match(r'(r|o)(\d+)', name)
        if match:
            prefix, number = match.groups()
            sheets_info.append((sheet.Name, int(number), prefix))  # Guardar nombre, número y prefijo
        else:
            sheets_info.append((sheet.Name, float('inf'), 'z'))  # Otras hojas al final

    # Ordenar por prefijo y número
    sheets_info.sort(key=lambda x: (x[2], x[1]))

    # Copiar hojas al libro nuevo usando nombres
    moved_count = 0
    for sheet_name, _, _ in sheets_info:
        try:
            sheet = original_wb.Sheets(sheet_name)  # Obtener hoja por nombre
            sheet.Copy(Before=new_wb.Sheets(new_wb.Sheets.Count))  # Copiar hoja al nuevo libro
            moved_count += 1
            print(f"Hojas copiadas: {moved_count}/{len(sheets_info)}")

            # Checkpoint: Guardar con un nombre único cada 30 hojas
            if moved_count % 30 == 0:
                checkpoint_path = os.path.join(save_dir, f"checkpoint_{moved_count}.xlsx")
                new_wb.SaveAs(checkpoint_path)
                print(f"Checkpoint guardado: {checkpoint_path}")

            time.sleep(0.2)  # Breve pausa para evitar sobrecargar Excel
        except Exception as e:
            print(f"Error al copiar la hoja '{sheet_name}': {e}")

    # Guardar el archivo nuevo al final
    final_path = os.path.join(save_dir, "Informe_Entregables_VF_prueba_final.xlsx")
    new_wb.SaveAs(final_path)
    print(f"Las hojas han sido copiadas y guardadas correctamente en: {final_path}")

except Exception as e:
    print(f"Se produjo un error general: {e}")

finally:
    # Cerrar ambos libros y Excel
    try:
        original_wb.Close(SaveChanges=False)
        new_wb.Close(SaveChanges=True)
    except:
        pass
    excel.Quit()
    print("Excel se ha cerrado correctamente.")
```

## Renombrar hojas

```python
# Iniciar Excel
excel = win32.Dispatch('Excel.Application')
excel.Visible = False  # Ejecutar Excel en segundo plano

# Abrir el libro de trabajo
wb = excel.Workbooks.Open(file_path)

# Contadores para fig y tab
fig_count = 1
tab_count = 1
save_interval = 20  # Guardar después de renombrar este número de hojas

# Renombrar hojas y mostrar progreso
print("Iniciando el renombrado de hojas...")

for i, sheet in enumerate(wb.Sheets):
    original_name = sheet.Name  # Nombre original de la hoja
    sheet_name = original_name.lower()  # Normalizar nombre en minúsculas

    # Renombrar según contenga "fig" o "tab"
    if "fig" in sheet_name:
        new_name = f'Fig{fig_count}'
        sheet.Name = new_name
        print(f"Renombrada hoja '{original_name}' a '{new_name}'")
        fig_count += 1
    elif "tab" in sheet_name:
        new_name = f'Tab{tab_count}'
        sheet.Name = new_name
        print(f"Renombrada hoja '{original_name}' a '{new_name}'")
        tab_count += 1

    # Guardar cada cierto número de hojas
    if (i + 1) % save_interval == 0:
        wb.Save()
        print(f"Progreso guardado después de renombrar {i + 1} hojas.")

# Guardar cambios finales
wb.Save()
wb.Close()
excel.Quit()

print("Renombrado completado y archivo guardado.")
```
